```python
import networkx as nx
import matplotlib.pyplot as plt
%pylab inline
```

```python
G=nx.Graph()
G.add_node('C')
G.add_nodes_from([1,2,3,4,5,6])
G.add_edge('C',1)
G.add_edges_from([('C',2),('C',3),('C',4),('C',5),('C',6)])
```

```python
print(G.nodes())
print(G.edges())
```

```python
pos={1:(1,0),2:(0.5,0.866),3:(-0.5,0.866),4:(-1,0),5:(-0.5,-0.866),6:(0.5,-0.866),'C':(0,0)}
nx.draw(G,pos=pos,with_labels=True)
```

```python
G2=nx.DiGraph()
G2.add_edges_from([('C',1),('C',2),('C',3),('C',4),('C',5),('C',6)])
```

```python
nx.draw(G2,pos=pos,with_labels=True,arrows=True)
```

```python
def make_graph(n):
    G4=nx.DiGraph()
    Edge=[]
    for i in range(n):
        Edge.append(('C',i))
    G4.add_edges_from(Edge)
    pos={'C':(0,0)}
    
    for i in range(n):
        xPos=cos(2*pi*(i-1)/n)
        yPos=sin(2*pi*(i-1)/n)
        pos[i]=(xPos,yPos)
    nx.draw(G4,pos=pos,with_labels=True,arrows=True)
```

```python
make_graph(20)
```
